## **Problem 1: Employee Performance Bonus Eligibility**

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find the highest score
highest_score = max(employees.values())

# Find all employees who have the highest score (handle ties)
top_performers = [name for name, score in employees.items() if score == highest_score]

# Display result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

## **Problem 2: Search Query Keyword Analysis**

```python
import string

query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Split into words
words = query.split()

# Count word frequency
frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Filter words searched more than once
repeated_keywords = {word: count for word, count in frequency.items() if count > 1}

print(repeated_keywords)
```

## **Problem 3: Sensor Data Validation**

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Store valid readings (even numbers)
valid_readings = [(index, value) for index, value in enumerate(sensor_readings) if value % 2 == 0]

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

## **Problem 4: Email Domain Usage Analysis**

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Extract domains and count
for email in emails:
    domain = email.split('@')[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total_users = len(emails)

# Calculate and display percentage usage
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

## **Problem 5: Sales Spike Detection**

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average sales
average_sales = sum(sales) / len(sales)

# Threshold for spike (30% above average)
threshold = average_sales * 1.3

# Detect spike days
for index, value in enumerate(sales):
    if value > threshold:
        print(f"Day {index + 1}: {value}")
```

## **Problem 6: Duplicate User ID Detection**

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

# Count occurrences
for user_id in user_ids:
    id_count[user_id] = id_count.get(user_id, 0) + 1

# Display duplicates only
for user_id, count in id_count.items():
    if count > 1:
        print(f"{user_id} → {count} times")
```
